```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
B= '\033[1m'
E= '\033[0m'
```

```python
#Import the CSV and display the header
data = pd.read_csv("creditcard.csv")
data.head()
```

```python
#Portray the frequency of fraud
count_classes = pd.value_counts(data['Class'], sort = True).sort_index()
count_classes.plot(kind = 'bar')
plt.title("Correct Vs. Fraud")
plt.xlabel("Class: {0} = Correct \n{1} = Fraud")
plt.ylabel("Samples")
plt.show()

correct = count_classes.tolist()[0]
fraud = count_classes.tolist()[1]
print(B+"Number of Correct Samples:"+E, correct)
print(B+"Number of Fraud Samples:"+E, fraud)
```

# As shown above the # of Correct samples greatly outweigh the # of Fraud samples. Thus the given dataset is imbalanced. Regular PCA feature engineering would not work as the  requires performance metrics to maximize precision.
